```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the credit card dataset
credit_card_data=pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')
```

```python
legit_sample=legit.sample(n=492)
```

```python
new_data_set = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_data_set.head()
```

```python
new_data_set.tail()
```

```python
# separating features and target
X=new_data_set.drop(columns='Class',axis=1)
Y=new_data_set['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# splitting data into training and testing sets
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model=LogisticRegression()
```

```python
# training Logistic Regression model
model.fit(X_train,Y_train)
```

```python
# predicting fraud or genuine transactions
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('accuracy on training data:',training_data_accuracy)
print('accuracy on test data:',test_data_accuracy)
```
